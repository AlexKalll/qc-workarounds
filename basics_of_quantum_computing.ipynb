```python
from qiskit_aer import Aer
```

```python
from qiskit import QuantumCircuit, transpile

# Create 2-qubit circuit with 2 classical bits for measurement
qc = QuantumCircuit(2, 2)

qc.h(0)
qc.x(1)
qc.cx(0, 1)
qc.z(0)
qc.y(1)
qc.measure([0, 1], [0, 1])
```

```python

# --- Simulation with 1000 shots ---
backend = Aer.get_backend('aer_simulator')
compiled_circuit = transpile(qc, backend)

# Run the simulation and get the result
job = backend.run(compiled_circuit)
result = job.result()
counts = result.get_counts()

# Show results
print("Measurement results (1000 samples):")
print(counts)

# Optional: draw circuit
qc.draw("mpl")
```

```python
qc_ = QuantumCircuit(3)

qc_.h(0)
for i in range(1, 3):
    qc_.cx(0, i)
qc_.draw("mpl")
```
