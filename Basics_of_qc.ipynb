##Quantum Computing vs Classical Computing

    Quantum Computing - Quantum Mechanics
    Classical Computing - Classical Physics

Classical computing adheres to determinstic systems where the same operations on a given classical bit always gives out the same results. However, in quantum computing , the system follows probabilistic nature and the results are the superposition of qubits.

##Quantum States and Systems

##State represents the description of a system at a given point of time.

```python
from qiskit.quantum_info import Statevector
import numpy as np
```

```python
#Matrix form representation for bits
ket0 = np.array([1, 0])
ket1 = np.array([0, 1])

print(ket0 / 2, ket1/ 2)
```

```python
#Operation on matrix representations
M1 = np.array([[1, 0],[0,1]])
M2 = np.array([[0, 1],[1 ,0]])

display(np.matmul(M1, ket0))
display(np.matmul(M2, ket0))
display(np.matmul(ket1, ket0))
```

```python
from qiskit.quantum_info import Statevector, Operator
import numpy as np
```

```python
zero, one = Statevector.from_label("0") , Statevector.from_label("1")
```

```python
display(zero)
display(one)
```

```python
zero.tensor(one).draw("latex")
```

```python
q_state1 = Statevector([1/np.sqrt(2) , 1/np.sqrt(2)])
q_state2 = Statevector([1/np.sqrt(2), 1j / np.sqrt(2)])
```

```python
q_state3 = q_state1.tensor(q_state2)
```

```python
q_state3.draw("latex")
```

```python
op1 = Operator([[1,0],[0,1]])
op2 = Operator([[0,1],[1,0]])
display(op1.draw("latex"))
display(op2.draw("latex"))
op1.tensor(op2).draw("latex")
```

```python
q_state3 = q_state3.evolve(op1 ^ op2)
```

```python
q_state3.draw("latex")
```

```python
q_state3.measure()
```

```python
W = Statevector([0, 1, 1, 0, 1,0,0,0]/np.sqrt(3))
```

```python
W.draw("latex")
```

```python
result, new_sv = W.measure([0])
```

```python
print(f"Result: {result} \nState after measurement: ", end=" ")
new_sv.draw("latex")
```
